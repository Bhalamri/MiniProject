```python
#Prepare Data for Machine Learning to create the models for prediction and classification of heart diseases
# Initialize data and target
target = df['target']
features = df.drop(['target'], axis = 1)
```

```python
# Split the data into training set and testing set
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 0)
```

```python
 # Create Train and evaluate model
def fit_eval_model(model, train_features, y_train, test_features, y_test):
    
    """
    Function: train and evaluate a machine learning classifier.
    Args:
      model: machine learning classifier
      train_features: train data extracted features
      y_train: train data lables
      test_features: train data extracted features
      y_test: train data lables
    Return:
      results(dictionary): a dictionary of classification report
    """
    results = {}
    
    # Train the model
    model.fit(train_features, y_train)
    
    # Test the model
    train_predicted = model.predict(train_features)
    test_predicted = model.predict(test_features)
    
     # Classification report and Confusion Matrix
    results['classification_report'] = classification_report(y_test, test_predicted)
    results['confusion_matrix'] = confusion_matrix(y_test, test_predicted)
        
    return results
```

```python
# Initialize the models
# Machine Learning models used for the prediction and classification of heart diseases:
# 1. Support Vector Machine(SVM)
# 2. Random Forest Classifier

sv = SVC(random_state = 1)
rf = RandomForestClassifier(random_state = 1)

# Fit and evaluate models
results = {}
for cls in [sv, rf]:
    cls_name = cls.__class__.__name__
    results[cls_name] = {}
    results[cls_name] = fit_eval_model(cls, X_train, y_train, X_test, y_test)
    
```

```python
# Print classifiers results report
for result in results:
    print (result)
    print()
    for i in results[result]:
        print (i, ':')
        print(results[result][i])
        print()
    print ('-----')
    print()
```

```python
# Evaluation the performance and accuracy using k-fold cross validation score
print("k fold Cross Val Score")
clf=SVC(kernel="rbf")
print("SVC={:.2f}%".format(np.mean(cross_val_score(clf,features,target,cv=10))*100))
rf_clf=RandomForestClassifier(n_estimators=10,criterion='entropy', random_state=0)
print("Random Forest Classifier={:.2f}%".format(np.mean(cross_val_score(rf_clf,features,target,cv=10))*100))
```
